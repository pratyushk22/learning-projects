```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
#creating a random artificial classification

X,y= make_classification(n_samples=200, n_informative=1, n_redundant=0, n_features=1, n_clusters_per_class=1)

X_train, X_test, y_train, y_test= train_test_split(X,y,stratify=y) #stratify is used for classification to preserve balance in classification

model= linear_model.LinearRegression()
model.fit(X_train,y_train)

y_pred= model.predict(X_test)
y_class= (y_pred>0.5).astype(int)

plt.scatter(X_test, y_test, color='b')
plt.scatter(X_test, y_class, color='r', marker='x')
plt.show()
```
